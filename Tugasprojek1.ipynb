# 1. Menginputkan 10 Nilai Mahasiswa

```python
arr = []
for i in range(10):
    angka = int(input("Masukkan Nilai: "))
    arr.append(angka)
```

Merupakan O(n), karena program melakukan input untuk setiap mahasiswa

# 2. Menampilkan Nilai Tertinggi, Terendah, dan Rata-rata

```python
print("\n Array:", arr)

print("\n Nilai Tertinggi:", max(arr))
print("\n Nilai Terendah:", min(arr))
print("\n Rata-rata:", sum(arr)/len(arr))
```

#

Merupakan O(n), karena program harus memeriksa seluruh array yang sudah di inputkan tadi

# 3. Menampilkan jumlah Mahasiswa yang lulus dan tidak lulus

```python
lulus = 0
for n in arr:
  if n >= 60:
    lulus += 1

tidak_lulus = len(arr) - lulus

print("\n Jumlah mahasiswa yang lulus: ", lulus)

print("\n Jumlah mahasiswa yang tidak lulus: ", tidak_lulus)
```

Merupakan O(n), karena semua nilai yang sudah di inputkan harus di cek
